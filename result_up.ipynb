```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://results.eci.gov.in/PcResultGenJune2024/partywiseresult-S24.htm'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find('table')

print(table)
```

```python
world_titles = table.find_all('th')

world_titles
```

```python
world_table_titles = [ title.text for title in world_titles ]

print(world_table_titles)
```

```python
del world_table_titles [4:8]

print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
column_data 
```

```python
for row in column_data[1:6]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    #print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
df
```

```python
df.loc[len(df)] = ['Total', '175', '0', '175'] 
df
```
